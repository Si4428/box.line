```python
import pandas as pd
ds = pd.read_csv("amazon.csv")
ds = ds.head(10)
ds.columns
```

```python
import matplotlib.pyplot as plt
plt.plot(ds['product_id'],ds['actual_price'], ds['discounted_price'])
plt.show()
```

```python

plt.bar(ds["actual_price"],ds['discounted_price'], color = 'y')

plt.xlabel("Actual price")
plt.ylabel("Discounted price")
plt.show()
```

```python

plt.boxplot(ds['actual_price'])
plt.show()
```

```python


plt.scatter(ds['product_id'], ds['discounted_price'])
plt.xlabel("Product ID")
plt.ylabel("Discounted Price")
plt.show()
```
